# LIVE ANALYSIS

```python
# import
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import datetime
```

```python
api_key = ""
api_secret = ""
```

```python
from binance.client import Client

def get_last_n_trades(symbol):
    client = Client(api_key, api_secret)
    orders = client.get_my_trades(symbol=symbol)
    return orders

def get_last_n_orders(symbol):
    client = Client(api_key, api_secret)
    orders = client.get_all_orders(symbol=symbol)
    return orders

def get_last_orders_for_all_symbols(all_symbols):
    orders_table = []
    for symbol in all_symbols:
        orders = get_last_n_orders(symbol)
        for order in orders:
            orders_table.append(order)
    
    return orders_table

def get_last_trades_for_all_symbols(all_symbols):
    orders_table = []
    for symbol in all_symbols:
        orders = get_last_n_trades(symbol)
        for order in orders:
            orders_table.append(order)
    
    return orders_table
```

#### GET LAST ORDER BINANCE


```python
orders = get_last_orders_for_all_symbols(all_symbols+'BUSD')
```

```python
trades = get_last_trades_for_all_symbols(all_symbols+'BUSD')
```

```python
df = pd.DataFrame(trades)

df["time"] = df["time"].apply(lambda x: datetime.datetime.fromtimestamp(x / 1000.0))

df = df.set_index("time")
df = df.sort_index()

mask = df.index >= pd.to_datetime('2023-02-24 00:00:00')
df = df[mask]

df
```
